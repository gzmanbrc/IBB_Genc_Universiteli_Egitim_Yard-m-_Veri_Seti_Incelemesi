```python
# Düzce Üniversitesi'ne ait verileri filtrele
df= pd.read_csv("C:/Users/gzman/Downloads/dam_occupancy.csv")
duzce_uni_df = df[df['university'].str.contains('Düzce Üniversitesi', case=False, na=False)]

# Düzce Üniversitesi verilerinin temel istatistikleri
duzce_uni_stats = duzce_uni_df.describe()

# Diğer üniversitelerle karşılaştırma
# Diğer üniversitelere ait verileri filtrele
other_unis_df = df[~df['university'].str.contains('Düzce Üniversitesi', case=False, na=False)]

# Diğer üniversitelerin temel istatistikleri
other_unis_stats = other_unis_df.describe()

# Karşılaştırma için birleştirme
comparison_stats = pd.concat([duzce_uni_stats, other_unis_stats], axis=1, keys=['Düzce Üniversitesi', 'Diğer Üniversiteler'])

# Sonuçları yazdırma
print(comparison_stats)

# İstatistiksel karşılaştırma grafikleri
import matplotlib.pyplot as plt

# Yaş dağılımı karşılaştırması
plt.figure(figsize=(12, 6))
plt.hist(duzce_uni_df['age'], bins=20, alpha=0.5, label='Düzce Üniversitesi')
plt.hist(other_unis_df['age'], bins=20, alpha=0.5, label='Diğer Üniversiteler')
plt.title('Yaş Dağılımı Karşılaştırması')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.legend(loc='upper right')
plt.show()

# Yardım dönemi karşılaştırması
plt.figure(figsize=(12, 6))
plt.hist(duzce_uni_df['assistance_term'], bins=20, alpha=0.5, label='Düzce Üniversitesi')
plt.hist(other_unis_df['assistance_term'], bins=20, alpha=0.5, label='Diğer Üniversiteler')
plt.title('Yardım Dönemi Karşılaştırması')
plt.xlabel('Yardım Dönemi')
plt.ylabel('Frekans')
plt.legend(loc='upper right')
plt.show()

# Öğrenci sayısı karşılaştırması
plt.figure(figsize=(12, 6))
plt.hist(duzce_uni_df['number_of_students'], bins=20, alpha=0.5, label='Düzce Üniversitesi')
plt.hist(other_unis_df['number_of_students'], bins=20, alpha=0.5, label='Diğer Üniversiteler')
plt.title('Öğrenci Sayısı Karşılaştırması')
plt.xlabel('Öğrenci Sayısı')
plt.ylabel('Frekans')
plt.legend(loc='upper right')
plt.show()
```
